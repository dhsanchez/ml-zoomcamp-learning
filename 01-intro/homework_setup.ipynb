 1. Select all cars from Asia
 2. Keep only 'vehicle_weight'  and 'model_year'

```python
asian_cars = df.loc[df.origin == 'Asia']
asian_cars_dropped = asian_cars[["vehicle_weight", "model_year"]]
asian_cars_dropped.head()
```

```python
# 3. Select the first 7 values as an numpy array
asian_cars_dropped[:7]

#Get the underlying NumPy array. Let's call it X.
import numpy as np

X= np.array(asian_cars_dropped[:7])
X
```

```python
# 4.Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.

XTX =  np.dot(X.T, X)

# 5.Invert XTX.
XTX_inv = np.linalg.inv(XTX)

# 7. Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
y= np.array([1100, 1300, 800, 900, 1000, 1100, 1200])

#8. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
w = XTX_inv @ X.T @ y
w
#9. What's the sum of all the elements of the result?
w.sum()
```
